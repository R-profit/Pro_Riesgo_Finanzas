```python
#Funciones numericas adicionales
import numpy as np

#lectura de datos y Manejo data-sects
import pandas as pd

#Api Datos
import yfinance as yfin

#graficos
import matplotlib.pyplot as plt

#velas japonesas
import mplfinance as mpf
```

```python
#descargar datos de 1año de la accion tesla

df = yfin.download('TSLA', start='2020-01-01', end='2021-01-01')
df
```

```python
#descargar 1 año de datos del S$P 500
df2 = yfin.download('^GSPC', start='2020-01-01', end='2021-01-01')
df2
```

```python
#grafico de tesla vs mercado
plt.figure(figsize=(16,8))
plt.title('Tesla VS Nercado')
plt.xlabel('Fecha')
plt.ylabel('Precios de cierre')
plt. plot(df['Close'], color='red')
plt. plot(df2['Close'], color='blue')
plt.legend(['Tesla','S&P500'],loc='lower right')
plt.show()
```

```python
#escalar las series a un valor de 100usd
df['Close_100'] = (100/df['Close'][0])*df['Close']
df2['Close_100'] = (100/df2['Close'][0])*df2['Close']
df
```

```python
#grafico de tesla vs mercado
plt.figure(figsize=(16,8))
plt.title('Tesla VS Nercado')
plt.xlabel('Fecha')
plt.ylabel('Precios de cierre escalados (2020-01-02: 100)')
plt. plot(df['Close_100'], color='red')
plt. plot(df2['Close_100'], color='blue')
plt.legend(['Tesla','S&P500'],loc='lower right')
plt.show()
```

```python
#diagrama de velas japonesas
mpf.plot(df[-42:],
         hlines = dict(hlines=[500, 600], colors=['red', 'green'], linestyle='-.'),
         type='candle',
         volume=True,
         figratio=(16,8),
         style='yahoo',
         title='velas japonesas Tesla')
```
