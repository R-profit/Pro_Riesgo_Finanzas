```python
#Api Datos
import yfinance as yfin

#Manejo de tablas de datos
import pandas as pd

#manejo de Archivos
from google.colab import files
#graficos
import matplotlib.pyplot as plt
```

```python
ticker = "GOOG"
df = yfin.download(ticker, start="2019-01-01", end="2021-01-01")
df
```

```python
#Guardar en .CSV
df.to_csv("GOOG.csv")

#descargar . csv
files.download("GOOG.csv")
```

```python
#cargar el archivo
files.upload()
```

```python
df2 = pd.read_csv("GOOG.csv")
df2['Date'] = pd.to_datetime(df2['Date'])
df2.set_index('Date',inplace=True)
df2
```

```python
# estadisticas describtivas
df2.describe()
```

```python
#grafico de los precios de cierre
plt.figure(figsize=(15,8))
plt. plot(df2['Close'], color = 'blue')
plt.title('Google [$GOOG]')
plt.xlabel('Fecha')
plt.ylabel('Precio de cierre')
plt.show()
```
