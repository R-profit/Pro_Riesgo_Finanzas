```python
#Funciones numericas adicionales
import numpy as np

#lectura de datos y Manejo data-sects
import pandas as pd


#Api Datos
import yfinance as yfin

#graficos
import matplotlib.pyplot as plt

#velas japonesas
import mplfinance as mpf

```

```python
#descargar datos de 1año de la accion CEMEX

df = yfin.download('CX', start='2020-01-01', end='2021-01-01')
df
```

```python
#dGrafico velas japonesas Cemex
mpf.plot(df,
         type='candle',
         volume=True,
         figratio=(16,8),
         style='yahoo',
         title='velas japonesas Cemex')
```

```python
#descargar datos de 1año de la accion ECOPETROL

df2 = yfin.download('EC', start='2020-01-01', end='2021-01-01')
df2
```

```python
#dGrafico velas japonesas Ecopetrol
mpf.plot(df2,
         type='candle',
         volume=True,
         figratio=(19,8),
         style='yahoo',
         title='velas japonesas Ecopetrol')
```

```python
#grafico filtrado Cemex
mpf.plot(df["2020-03-01":"2020-04-01"],
         type='candle',
         volume=True,
         figratio=(19,8),
         style='yahoo',
         title='CEMEX marzo-abril')
```

```python
df3= yfin.download('EC',start='2022-04-04', end='2022-04-09',
interval="1h")
df3
```

```python
#grafico filtrado una semana Ecopetrol
mpf.plot(df3,
         type='candle',
         volume=True,
         figratio=(19,8),
         style='yahoo',
         title='ECOPETROL 1 semana',
         mav=(5, 20))
```
